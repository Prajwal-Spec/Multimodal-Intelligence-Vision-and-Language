```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image
import requests
```

```python
url = 'http://images.cocodataset.org/val2017/000000039769.jpg'
image = Image.open(requests.get(url, stream=True).raw)
```

```python
feature_extracter = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
inputs = feature_extracter(images=image, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
```

```python
# model predicts one of the 1000 imagenet classes
```

```python
predicted_class_idx = logits.argmax(-1).item()
```

```python
print("Predicted Class: ",predicted_class_idx, ": ", model.config.id2label[predicted_class_idx])
```
